Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
#Retrieiving all the columns from the master table
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
```

```python
# your code goes here
df['ConvertedComp'].plot(kind='hist', bins=5, edgecolor='black', align='left')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df['Age'].plot(kind='box')
plt.ylabel('Values')
plt.title('Age Box Plot')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
df['WorkWeekHrs'].mean()
```

```python
# your code goes here
df.plot.scatter(x='Age', y='WorkWeekHrs')
plt.title('Correlation between Age and Work week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(10, 6))  
plt.scatter(
    x='WorkWeekHrs',
    y='CodeRevHrs',
    s=df['Age'] * 10,  #'*10' is for scaling the size of the bubbles
    data=df,
    alpha=0.5 #alpha is for the transcperency 
)

# Adding labels and title
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')

# Displaying the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
#Retrieiving all the columns from the master table
QUERY = """
SELECT *
FROM DatabaseDesireNextYear;
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_db = pd.read_sql_query(QUERY,conn)
```

```python
#print the first 5 of the new dataframe
top_databases = df_db['DatabaseDesireNextYear'].value_counts().head(5)
top_databases
```

```python
# your code goes here
plt.figure(figsize=(8, 6))  # Set the figure size
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Displaying the pie chart
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
filtered_data = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculating median WorkWeekHrs and CodeRevHrs for the filtered data
median_values = filtered_data.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Plotting the stacked chart
median_values.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')

# Displaying the chart
plt.legend(title='Metrics')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
filtered_data = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

median_values = filtered_data.groupby('Age')[['ConvertedComp']].median()

#plotting the graph
median_values.plot(kind='line', figsize=(10, 6))
plt.title('Median Converted Comp from ages 45-60')
plt.xlabel('Ages')
plt.ylabel('Converted Comp')
plt.show()
```
